#### Exercise 1 - Clustering

##### Bisecting K-Means

This first cell provides all the neccessary functions for completing
the divisive K-means algorithm. 

```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances

def SSE(X, centroid):
    """
    Return the Sum of Sum of squared errors
    """
    return sum(sum((centroid - X) ** 2)) / 2


def pickNextSplit(clusters: list):
    """
    Compares the SSE of clusters in a list of clusters and picks the one with the highest SSE for splitting
    by popping it from the list.
    """

    previous_sse = -1  # initial value

    for idx, cluster in enumerate(clusters):
        cluster_temp = cluster[0]  # unpack
        current = SSE(cluster_temp[:, :1], cluster[1])

        if current > previous_sse:
            previous_sse = current  # new high
            highest = idx

    largest_cluster = clusters.pop(highest)
    return largest_cluster, previous_sse


def filterClusters(X, pred):
    # Concat prediction with data points for filtering
    copy = np.copy(X)
    concat = np.c_[copy[:, 0], copy[:, 1], pred]

    # Split clusters
    zeroes = concat[concat[:, 2] == 0]
    ones = concat[concat[:, 2] == 1]
    return zeroes, ones,


def bestSplit(Xn, iter, saved_clusters):
    """
    Split iter times and return the best splits and the next split.
    """
    SSE_best = -1

    for it in range(iter):
        temp = saved_clusters.copy()

        # Fit and predict clusters
        clf = KMeans(n_clusters=2)
        clf.fit(Xn)
        pred = clf.predict(Xn)  # Best predictions after n iterations of 2 clusters

        # Split based on predictions
        cl_0, cl_1 = filterClusters(Xn, pred)

        # Add splits and their centers to list of clusters
        temp.append((cl_0, clf.cluster_centers_[0, :]))
        temp.append((cl_1, clf.cluster_centers_[1, :]))

        # Pick the next split
        split, SSE_current = pickNextSplit(temp)

        # Keep clusters for next split
        if SSE_current > SSE_best or SSE_best == -1:
            SSE_best = SSE_current  # replace best sse with current
            next_split = split  # save for split
            clusters = temp

    #print(f"Next split: {SSE_best}")

    return next_split, clusters


def recursiveSplit(Xn, k_clusters, iterations, saved_clusters):

    nextSplit, clusters = bestSplit(Xn, iterations, saved_clusters)

    # Add (cluster, center) to saved clusters
    saved_clusters = clusters.copy()

    if k_clusters == 1 or nextSplit[0].shape[0] < 2:
        # Append remainder of cluster
        saved_clusters.append(nextSplit)
        return saved_clusters

    return recursiveSplit(nextSplit[0], k_clusters - 1, iterations, saved_clusters)


def mergeClusters(clusters):
    # Pop first cluster the initial group
    merged = np.array(clusters.pop(0)[0])
    merged[:, 2] = 0

    lastIdx = len(clusters)-1
    last = clusters.pop()
    last[0][:, 2] = lastIdx  # merge remainder with last cluster

    # Merge the rest
    for idx, cluster in enumerate(clusters):
        cluster[0][:, 2] = idx + 1  # set index of cluster
        next = cluster[0]
        merged = np.append(merged, next, axis=0)

    # merge last cluster manually
    merged = np.append(merged, last[0], axis=0)

    return merged


def sorter(A, B):
    """
    Function by Rafael Martins

    Uses three sorts. Two to align A and B. And one sort to fit B back into the order of A.

    """
    index = np.arange(A.shape[0])[A[:, 0].argsort()].argsort()
    return B[B[:, 0].argsort()][index]


def bkmeans(Xn, k_clusters, iterations):
    original = Xn.copy()

    saved_clusters = []  # no clusters from the start
    saved_clusters = recursiveSplit(Xn, k_clusters, iterations, saved_clusters)

    merged = mergeClusters(saved_clusters)

    merged = sorter(original, merged)

    return merged[:, 2]
```

##### Exercise 2. Sammon Mapping

The implementation of sammons mapping works by computing minimizing
the stress function known as Sammons Stress using a
gradient descent approach with partial derivatives.

I have noticed that the implementation is quite sensitive for large values
of alpha (learning rate), so the best approach is to use many iterations with low alpha which takes quite a while
since the function is very slow unfortunately.

```python
def sammon(X, iter, threshold, alpha):
    np.random.seed(10)
    y = np.random.rand(X.shape[0], 2)

    dist_x = pairwise_distances(X, X, 'euclidean')
    dist_x_sum = np.sum(dist_x) / 2
    counter = 0
    previous = -1
    c = np.divide(-2, dist_x_sum)

    for it in range(0, iter):

        # Calc distances
        dist_y = pairwise_distances(y, y, 'euclidean')
        dist_y_sum = np.sum(dist_y) / 2

        # Calc stress
        E = 1 / dist_x_sum * np.sum(np.divide(np.power(np.subtract(dist_y_sum, dist_x_sum), 2), dist_x_sum))
        # print(f"E({it + 1}): {E}")

        if E < threshold:
            print("Threshold exceeded! Stopping.")
            return y

        if E < previous:
            counter = 0
        else:
            counter += 1

        if counter > 3:  # Three consecutive increases in stress
            print("Convergence exploded. Returning current result..")
            return y
        previous = E

        for i in range(0, X.shape[0]):

            # Create a full array of y[i] to calc all y[i] - y[j] at the same time in an attempt to speed up process
            diff = np.full((y.shape), y[i]) - y

            # Initialise sums
            summ0 = np.array([0, 0])
            summ1 = np.array([0, 0])

            for j in range(i + 1, X.shape[0]):

                # First partial
                numerator = dist_x[i, j] - dist_y[i, j]
                denominator = np.multiply(dist_x[i, j], dist_y[i, j])

                if denominator < 1e-9:  # clamp denominator
                    denominator = 1e-9

                summ0 = summ0 + np.multiply(np.divide(numerator, denominator), diff[j])

                # Second partial
                p0 = np.divide(1, denominator)
                p1 = np.divide(np.power(diff[j], 2), dist_y[i, j])
                p2 = 1 + np.divide(numerator, dist_y[i, j])
                summ1 = summ1 + (p0 * (numerator - p1 * p2))

            partial0 = (c * summ0)
            partial1 = np.abs(c * summ1)

            delta = np.divide(partial0, partial1, where=partial1 != 0)
            y[i] -= alpha * delta
    return y
```

##### Exercise 3. Visualization of results

Data sets:

* MNIST - (600, 64). 9 classes: (0.9).

* South African Congenital Heart disease. (472, 9). Classes: positive/negative.

* Wine - (178, 13) 3 classes: (Class_0, class_1, class_2)

##### 3.1.1

Overall I think tSNE performed best on all data, it managed to separate all
data sets quite well compared to PCA and Sammon. 

Sammon seems to have a slight advantage over TSNE on MNIST (row 1) due to the more distinct separation of clusters. For row
two all algorithms struggled to separate the data where PCA and Sammon were pretty much equal and TSNE a bit better.

For the wine set TSNE separated the clusters well while Sammon and PCA struggled.
 
##### 3.1.2

The MNIST and Wine classes seem to be well suited for separation while the True/Positive classes of the SA Congenital Heart Disease set seems much harder to distinguish between. For the heart set the difficulty to separate the data might be due to small variances and dissimilarities between positive and negative which will affect the outcome of PCA and Sammon negatively.



```python
import numpy as np
from sklearn import preprocessing as pp, datasets
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

heart = np.genfromtxt('data/sa_heart.csv', delimiter=',', dtype=float, skip_header=True, usecols=(0,1,2,3,4,5,6,7,8))
digits = datasets.load_digits()
wine = datasets.load_wine()

X_digits = pp.StandardScaler().fit_transform(digits.data[:200, 1:])  # 25, 0.02, 9 classes
X_heart = pp.StandardScaler().fit_transform(heart[:200, :])  # 100, 0.002
X_wine = pp.StandardScaler().fit_transform(wine.data[:200, :])  # 100, 0.003 3 classes

pca = PCA(n_components=2)
tsne = TSNE(n_components=2)

# Fit
Xn1 = pca.fit_transform(X_digits)
Xn2 = sammon(X_digits, 500, 1e-6, 0.002)
Xn3 = tsne.fit_transform(X_digits)

print("Fit 1 complete")
Xn4 = pca.fit_transform(X_heart)
Xn5 = sammon(X_heart, 500, 1e-6, 0.002)
Xn6 = tsne.fit_transform(X_heart)
print("Fit 2 complete")
Xn7 = pca.fit_transform(X_wine)
Xn8 = sammon(X_wine, 150, 1e-6, 0.003)
Xn9 = tsne.fit_transform(X_wine)
print("Fit 3 complete")

# Digits
pred1 = bkmeans(Xn1, 9, 3)
pred2 = bkmeans(Xn2, 9, 3)
pred3 = bkmeans(Xn3, 9, 3)

# South African Congenital heart disease
pred4 = bkmeans(Xn4, 2, 3)
pred5 = bkmeans(Xn5, 2, 3)
pred6 = bkmeans(Xn6, 2, 3)

# Wine
pred7 = bkmeans(Xn7, 3, 3)
pred8 = bkmeans(Xn8, 3, 3)
pred9 = bkmeans(Xn9, 3, 3)

# Plots
color = ['red', 'green', 'blue', 'yellow', 'black', 'orange', 'pink', '#37AB65', '#3DF735',
         '#AD6D70', '#EC2504', '#8C0B90', '#C0E4FF', '#27B502', '#7C60A8', '#CF95D7', '#145JKH']

fig, ax = plt.subplots(nrows=3, ncols=3, figsize=(15,15))

def scatter(X, y,clusters, plotNr, title, legend=None):
    for n in range(0, clusters):
        ax = plt.subplot(3, 3, plotNr)
        filter = X[y == n]
        ax.set_title(title)
        ax.scatter(filter[:, 0], filter[:, 1], color=color[n], marker='o', s=1)

scatter(Xn1, pred1, 9, 1, "PCA")
scatter(Xn2, pred2, 9, 2, "Sammon")
scatter(Xn3, pred3, 9, 3, "t-SNE")

scatter(Xn4, pred4, 2, 4, "PCA")
scatter(Xn5, pred5, 2, 5, "Sammon")
scatter(Xn6, pred6, 2, 6, "t-SNE")

scatter(Xn7, pred7, 3, 7, "PCA")
scatter(Xn8, pred8, 3, 8, "Sammon")
scatter(Xn9, pred9, 3, 9, "t-SNE")

plt.show()
```

#### Exercise 3.2

This exercise will visualize t-SNE together with classical Kmeans, Hierarchical and Bisecting K means and compare results.


```python
from sklearn.cluster import KMeans, AgglomerativeClustering

tsne = TSNE(n_components=2)

Xn1 = tsne.fit_transform(X_digits)
Xn2 = tsne.fit_transform(X_heart)
Xn3 = tsne.fit_transform(X_wine)

kmeans = KMeans(n_clusters=9).fit(Xn1)
pred1 = kmeans.predict(Xn1)
pred2 = AgglomerativeClustering(n_clusters=9).fit_predict(Xn1)
pred3 = bkmeans(Xn1, 9, 5)

kmeans = KMeans(n_clusters=2).fit(Xn2)
pred4 = kmeans.predict(Xn2)
pred5 = AgglomerativeClustering(n_clusters=2).fit_predict(Xn2)
pred6 = bkmeans(Xn2, 2, 5)

kmeans = KMeans(n_clusters=3).fit(Xn3)
pred7 = kmeans.predict(Xn3)
pred8 = AgglomerativeClustering(n_clusters=3).fit_predict(Xn3)
pred9 = bkmeans(Xn3, 3, 5)
```

### Visualization of results

Hierarchical and Kmeans performed very similarly on all data sets while Bisecting Kmeans struggled on the MNIST (row1) and (Wine set row 3).

The reason for the differences I believe is the top-down approach. The agglomerative will merge the clusters until it has the amount of clusters specified while the divisive will split until x clusters have been found. 

This means that the last split (for uneven cluster targets) will contain a remainder of points which must be added to the last cluster and I think that it is the largest source of errors in the plots. This idea is supported by the results of row 2 where Bisecting Kmeans performs equally to the others. It might also be that the Bisection implementation is a little bit flawed.


```python
fig, ax = plt.subplots(nrows=3, ncols=3, figsize=(10,10))
def scatter(X, y,clusters, plotNr, title):
    for n in range(0, clusters):
        ax = plt.subplot(3, 3, plotNr)
        filter = X[y == n]
        ax.set_title(title)
        ax.scatter(filter[:, 0], filter[:, 1], color=color[n], marker='o', s=1)

scatter(Xn1, pred1, 9, 1, "KMeans")
scatter(Xn1, pred2, 9, 2, "Hierarchical")
scatter(Xn1, pred3, 9, 3, "Bisect KMeans")

scatter(Xn2, pred4, 2, 4, "KMeans")
scatter(Xn2, pred5, 2, 5, "Hierarchical")
scatter(Xn2, pred6, 2, 6, "Bisect KMeans")

scatter(Xn3, pred7, 3, 7, "KMeans")
scatter(Xn3, pred8, 3, 8, "Hierarchical")
scatter(Xn3, pred9, 3, 9, "Bisect KMeans")

plt.show()
```
